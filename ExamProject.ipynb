# 1. Convert the app size to a number

```python
size_int_re = re.compile('^(?P<number>\d+\.?\d*)(?P<unit>\w*)$')

def to_mult(unit):
    if (unit == 'G'):
        return 1000000000
    if (unit == 'M'):
        return 1000000
    if (unit == 'k'):
        return 1000
    return 1

def size_to_num(size):
    num = size_int_re.match(size)
    if num:
        unit = num.group('unit')
        mult = to_mult(unit)
        size = float(num.group('number')) * mult
    return size
```

```python
apps['Size'] = apps['Size'].apply(size_to_num)
apps.head()
```

# 4. Convert Current Ver and Android Ver into a dotted number

```python
ver_re = re.compile('^(?P<Ver>\d\.\d\.?\d?)(?P<Stringa>\w*)')

def version(ver):
    r = ver_re.match(str(ver))
    if r:
        ver = r.group('Ver')
    return(ver)
```

```python
apps['Android_Ver_Clean'] = apps['Android Ver'].apply(version)
apps[['Android Ver','Android_Ver_Clean']].head()
```

```python
apps['Current_Ver_Clean'] = apps['Current Ver'].apply(version)
apps[['Current Ver','Current_Ver_Clean']].head()
```

# 8. Create two dataframes: one for the genres and one bridging apps and genders. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
app_list = []
genres_list = []

for index,row in apps.iterrows():
    appo = row['Genres'].split(';')
    for genre in appo:
        genres_list.append(genre)
        app_list.append(row['App'])


app_genre = pd.DataFrame({'App' : app_list, 'Genres' : genres_list})
genres_df = pd.DataFrame({'Genres' : list(set(genres_list))})
```

# 9. For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
appo = pd.DataFrame(columns=list(set(genres_list)))
apps_copy = apps
apps_2 = apps_copy.join(appo)    

for index,row in apps_2.iterrows():
    for genres in list(set(genres_list)):
        apps_2.at[index, genres] = False
    x = row['Genres'].split(';')
    for genre in x:
        apps_2.at[index, genre] = True
```

# 11. For each app, compute the approximate income, obtain as a product of number of installs and price.

```python
re_price = re.compile('\d')

def p_to_n(price):
    price = price.strip('$')
    if (re_price.match(price)):
        price = float(price)
    else:
        price = np.nan
    return(price)
```

```python
apps['Price_value'] = apps['Price'].apply(p_to_n)

apps['Approx_Income'] = apps['Price_value'] * apps['Installs']

apps[apps['Price'] != '0'][['Price','Price_value','Installs','Approx_Income']].head()
```
